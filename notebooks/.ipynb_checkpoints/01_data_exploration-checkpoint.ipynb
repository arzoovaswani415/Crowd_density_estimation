```python
import os
import cv2
import h5py
import numpy as np
import matplotlib.pyplot as plt
import random

img_dir = "../data/ShanghaiTech/part_A/train_data/images"
gt_dir  = "../data/ShanghaiTech/part_A/train_data/ground-truth-h5"

img_list = os.listdir(img_dir)

for i in range(3):
    img_name = random.choice(img_list)
    img_path = os.path.join(img_dir, img_name)

    # load image
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # load density map
    gt_name = img_name.replace(".jpg", ".h5")
    gt_path = os.path.join(gt_dir, gt_name)

    with h5py.File(gt_path, 'r') as f:
        density = np.array(f['density'])

    print(f"Image: {img_name}")
    print("Estimated count from GT:", density.sum())
    print("Density min/max:", density.min(), density.max())

    # visualize
    plt.figure(figsize=(12,4))

    plt.subplot(1,2,1)
    plt.imshow(img)
    plt.title("Input Image")
    plt.axis("off")

    plt.subplot(1,2,2)
    plt.imshow(density, cmap='jet')
    plt.title("Ground Truth Density Map")
    plt.axis("off")
   # plt.savefig(f"../results/{img_name}_density.png")


    plt.show()
```
